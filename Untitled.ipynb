```python
import glob
import numpy as np
from math import floor
from scipy.signal import convolve2d
from PIL import Image
import matplotlib.style
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline

# check for latex in the host env
import subprocess
try:
    subprocess.check_call(['latex', '-help'])
except Exception as e:
    print(e)

#mpl.use("pgf")
mpl.rcParams.update({
    'figure.figsize': [8.0, 6.0],
    'figure.dpi' : 80,
    'savefig.dpi' : 80,
    "pgf.texsystem": "pdflatex",
    'font.family': 'serif',
    'text.usetex': True,
    'pgf.rcfonts': False,
})

def plot_image(image):
    fig = plt.figure()
    if len(image.shape) == 2:
        plt.imshow(image, cmap='gray')#, vmin=0, vmax=255)
    else:
        plt.imshow(image)
    plt.show()
    return fig
    
def save_figure(fig):
    plt.savefig(f"plot/{len(glob.glob('plot/*')):0>2}.pgf")

    
class Augmentation:
    
    @staticmethod
    def identity(x : np.ndarray) -> np.ndarray:
        '''
        (a) Original 
        '''
        return np.copy(x)
    
    @staticmethod
    def _get_zoomed_box(height: int, width: int, zoom_min : int = 0.2, zoom_max : int = 0.7, ratio : bool = True):
        '''
        return 
        '''
         # randomly select a coef within [zoom_min, zoom_max]
        zoom_coef = zoom_min + (zoom_max - zoom_min) * np.random.rand()
        #print(zoom_coef)
        
        # compute the new width and height
        if ratio:
            new_width = width * zoom_coef
            new_height = new_width * (height / width) 
        else:
            tmp = zoom_coef * min(width, height)
            new_width = tmp
            new_height = tmp 
        #print(new_width, new_height)
        
        # find a random x0 et y0
        x0 = (width-new_width) * np.random.rand()
        y0 = (height-new_height) * np.random.rand()
        
        x1 = x0 + new_width
        y1 = y0 + new_height
        #print(np.floor(x0), x1, y0, y1)
        
        return floor(x0), floor(x1), floor(y0), floor(y1)
        
    
    @staticmethod
    def crop(x : np.ndarray, zoom_min : int = 0.2, zoom_max : int = 0.7) -> np.ndarray:
        '''
        (b) Crop
        '''
        height, width = x.shape[:2]
        if len(x.shape) > 2:
            channel =  x.shape[2]
        #print(f"height={height} width={width}")
        
        x0, x1, y0, y1 = Augmentation._get_zoomed_box(width, height, zoom_min, zoom_max)
        
        out = np.copy(x[y0:y1, x0:x1, :])
        return out
    
    @staticmethod
    def color_distort(x : np.ndarray, zoom_min : int = 0.2, zoom_max : int = 0.7) -> np.ndarray:
        return x
    
    
    @staticmethod
    def flip(x : np.ndarray, axis : int = 0) -> np.ndarray:
        '''
        flip an image (default (axis=0) to vertically)
        '''
        assert axis < len(x.shape)
        out = np.copy(x)
        return np.flip(out, axis=axis)
        
    
    @staticmethod
    def flip_vertically(x : np.ndarray) -> np.ndarray:
        '''
        vertically (axis=0)
        '''
        return Augmentation.flip(x, axis=0)
    
    @staticmethod
    def flip_horizontally(x : np.ndarray) -> np.ndarray:
        '''
        horizontally (axis=1)
        '''
        return Augmentation.flip(x, axis=1)
    
    @staticmethod
    def cutout(x : np.ndarray, zoom_min : int = 0.2, zoom_max : int = 0.7) -> np.ndarray:
        '''
        set a image's random portion to gray
        '''
        height, width = x.shape[:2]
        if len(x.shape) > 2:
            channel =  x.shape[2]
            
        x0, x1, y0, y1 = Augmentation._get_zoomed_box(width, height, zoom_min, zoom_max, ratio=False)
        
        out = np.copy(x)
        out[y0:y1, x0:x1, :] = 149
        
        return out
    
    @staticmethod
    def rotate_90(x : np.ndarray, multiple : int = -1):
        '''
        rotate an image 
            0 :   0 
            1 :  90 
            2 : 180 
            3 : 270 
        '''
        if multiple == -1: # draw a random angle index
            multiple = np.random.randint(5)
        if multiple == 0:
            out = np.copy(x)
        else:
            out = np.copy(np.rot90(x, k=multiple%5, axes=(0, 1)))
        return out
    
    @staticmethod
    def gaussian_noise(x : np.ndarray, mean : float = 0.0, var : float = 200.) -> np.ndarray:
        '''
        add a gaussian noise to the image
        '''
        sigma = var ** 0.5 # std
        noise = np.random.normal(loc=mean, scale=sigma, size=x.shape)
        out = np.empty(x.shape, dtype=np.uint8)
        out[:] = np.clip(x + noise, 0, 255)
        return out
    
    @staticmethod
    def salt_pepper_noise(x : np.ndarray, amount : float = 0.02, ratio : float = 0.5) -> np.ndarray:
        '''
        add a salt and pepper noise to the image
        '''
        out = np.copy(x)
        row, col, _ = x.shape
    
        
        num_salt = np.ceil(amount * x.size * ratio)
        num_pepper = np.ceil(amount * x.size * (1.0 - ratio))

        # Add Salt noise
        coords = [np.random.randint(0, i - 1, int(num_salt)) for i in x.shape]
        out[coords[0], coords[1], :] = 255

        # Add Pepper noise
        coords = [np.random.randint(0, i - 1, int(num_pepper)) for i in x.shape]
        out[coords[0], coords[1], :] = 0
        
        return out

    
    @staticmethod
    def _filter(x : np.ndarray, kernel : np.ndarray) -> np.ndarray:
        '''
        filter the image x with the kernel
        '''
        if len(x.shape) == 2: # grayscale
            out = convolve2d(x, kernel, mode='same', boundary='symm')
        elif len(x.shape) == 3 and x.shape[2] == 3:  # rgb
            out = np.empty(x.shape, dtype=np.uint8)
            for c in range(3):
                out[:, :, c] = convolve2d(x[:, :, c], kernel, mode='same', boundary='symm')
        out[:] = np.clip(out, 0, 255)
        return out
    
    @staticmethod
    def gaussian_blur(x : np.ndarray) -> np.ndarray:
        '''
        blur image with gaussian
        '''
        gaussian_kernel = np.array([[1.0, 2.0, 1.0], [2.0, 4.0, 2.0], [1.0, 2.0, 1.0]]) # 3x3 kernel
        gaussian_kernel /= np.sum(gaussian_kernel)
        return Augmentation._filter(x, gaussian_kernel)
    
    @staticmethod
    def _grayscale(x : np.ndarray) -> np.ndarray:
        if len(x.shape) == 3:
            return 0.2989 * x[:,:,0] + 0.5870 * x[:,:,1] + 0.1140 * x[:,:,2]
        else:
            return x
    
    @staticmethod
    def sobel_filter(x : np.ndarray) -> np.ndarray:
        '''
        blur image with gaussian
        '''
        Gx = np.array([
            [-1.,  0., +1.], 
            [-2.,  0., +2.], 
            [-1.,  0., +1.]
        ]) 
        Gy = np.array([
            [-1., -2., -1.], 
            [ 0.,  0.,  0.], 
            [+1., +2., +1.]
        ])
        if len(x.shape)==2:
            return np.sqrt(Augmentation._filter(x, Gx)**2 + Augmentation._filter(x, Gy)**2)
        elif len(x.shape)==3:
            x_grayscale = Augmentation._grayscale(x)
            x_dx = Augmentation._filter(x_grayscale, Gx)
            x_dy = Augmentation._filter(x_grayscale, Gy)
            return np.sqrt(x_dx*x_dx + x_dy*x_dy)

    @staticmethod    
    def scharr_filter(x : np.ndarray) -> np.ndarray:
        scharr = np.array([
            [-3 -3j, 0 -10j,  +3 -3j],
            [-10+0j, 0 + 0j, +10 +0j],
            [-3 +3j, 0 +10j,  +3 +3j]
        ])
        if len(x.shape)==2:
            return Augmentation._filter(x, scharr)
        elif len(x.shape)==3:
            x_grayscale = Augmentation._grayscale(x)
            grad = Augmentation._filter(x_grayscale, scharr)
            return np.absolute(grad)
        
#load image
im = Image.open('data/dog_von.grzanka.png')
I = np.asarray(im)
im = Image.fromarray(I)


fig = plt.figure()
row, col = 5, 2

fig, axs = plt.subplots(row, col, figsize=(6, 12))
axs = axs.flatten()

X = [
    Augmentation.crop(I),
    Augmentation.flip_vertically(I),
    Augmentation.flip_horizontally(I),
    Augmentation.cutout(I),
    Augmentation.rotate_90(I),
    Augmentation.salt_pepper_noise(I),
    Augmentation.gaussian_noise(I),
    Augmentation.gaussian_blur(I),
    Augmentation.sobel_filter(I),
    Augmentation.scharr_filter(I)
]
T = [
    "crop",
    "flip vertically",
    "flip horizontally",
    "cutout",
    "rotate 90",
    "salt-pepper noise",
    "gaussian noise",
    "gaussian blur",
    "sobel filter",
    "scharr filter"
]
for ax, x, t in zip(axs, X, T):
    ax.set_title(t)
    ax.imshow(x, cmap='gray' if len(x.shape)==2 else None)
plt.subplots_adjust(wspace=0.5, hspace=0.5)
plt.show()
print('ok')
```
